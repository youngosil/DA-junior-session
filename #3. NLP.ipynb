**Subword Tokenization**
---
1. Byte Pair Encoding(BPE) : BPE는 1994년에 제안된 데이터 압축 방식 중 하나인데, 이를 텍스트 데이터에 적용하여 단어를 분할하는 방식으로 사용됨. BPE에서는 말뭉치(corpus)에서 가장 자주 등장하는 서브워드를 병합하여 더 큰 서브워드를 생성하는 과정을 반복하면서, 최종적으로 원하는 단어 수준의 서브워드 집합을 생성. BPE는 자주 등장하는 서브워드를 합치므로서 데이터의 압축률을 높일 수 있음
---
2. Unigram Language Model: Unigram Language Model(유니그램 언어 모델)은 텍스트 데이터의 서브워드 생성 방식 중 하나. 이 모델은 입력된 텍스트를 공백으로 구분된 단어가 아닌 문자(Character) 단위로 분할하며, 해당 문자들의 빈도수를 기반으로 서브워드 집합을 생성. Unigram Language Model은 텍스트 데이터의 언어 모델링에 있어서 간단하면서도 효과적인 방법 중 하나이지만 서브워드가 단어의 의미를 제대로 파악하지 못할 가능성이 있으며, 다른 Subword Tokenization 방식에 비해 정확도가 상대적으로 낮을 수 있음. + Unigram Language Model의 동작 방식은 다음과 같음.
- 입력 텍스트 데이터를 모든 문자(Character) 단위로 분할
- 분할된 문자들의 빈도수를 세어, 상위 N개의 문자를 서브워드로 선택, 이 때 N은 미리 정해진 하이퍼파라미터로, 주로 8,000~32,000 정도로 설정
- 선택된 서브워드 집합을 기반으로, 텍스트 데이터를 서브워드로 분할. 이 때, 서브워드는 문자(Character) 단위로 분할된 문자열의 N-gram으로 생성. 예를 들어, 'hello' 문자열을 N=3으로 설정하면, ['hel', 'ell', 'llo']의 세 개의 서브워드로 분할됨
- 서브워드 분할이 완료된 텍스트 데이터를 토크나이즈하여 모델링에 사용
---
3. Sentencepiece (Google) :`pip install sentencepiece` 로 설치 가능, PE와 Unigram Language Model 등의 서브워드 생성 방식을 모두 지원하며, 다양한 언어 모델에 적용할 수 있는 범용적인 토크나이저로 자리 잡고 있음. SentencePiece는 훈련 데이터로부터 서브워드 집합을 생성하는데, 입력 텍스트의 문자열을 일정한 길이의 서브워드로 분할하고, 이를 다시 새로운 서브워드로 합치면서 단어 분할을 수행함.  SentencePiece는 문자 단위(subword)로 분할하기 때문에 단어의 의미를 잘 보존하면서도, 학습 데이터의 언어적 특성을 반영할 수 있습니다. 또한, 새로운 단어가 등장하더라도 해당 단어를 서브워드로 나눌 수 있으므로, 일반적인 언어 모델보다 높은 범용성을 보임.SentencePiece는 C++로 구현되어 있으며, Python, Java, Ruby, C#, Go 등 다양한 프로그래밍 언어에서 사용 가능함. TensorFlow, PyTorch, Apache MXNet 등의 딥러닝 프레임워크와도 호환되어, 텍스트 데이터 처리에 있어서 매우 유용한 도구 중 하나.

```python
from transformers import pipeline

pipe = pipeline("text-classification")
pipe("Daewoo Hall is too far from the main gate")
```

```python
from transformers import pipeline

models = ["bert-base-uncased", "roberta-base", "distilbert-base-uncased", "google/electra-small-discriminator", 
          "albert-base-v2", "xlm-roberta-base"]
sentences = ["Daewoo Hall is too far from the main gate"]

for model_name in models:
    print(f"Model: {model_name}")
    classifier = pipeline("text-classification", model=model_name)
    result = classifier(sentences)
    print(result)
    print()
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
text = "Daewoo Hall is too far from the main gate"
tokenized_text = tokenizer.tokenize(text)

print(tokenized_text)
```

```python
from transformers import pipeline, AutoTokenizer

model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
classifier = pipeline('text-classification', model=model_name)

result = classifier(tokenized_text)
```
