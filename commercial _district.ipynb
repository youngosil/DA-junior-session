```python
print(data)
```

```python
columns = data['header']['columns']
print(columns)
```

```python
# CSV 파일 저장
with open("commercial_district.csv", "w", encoding="utf-8-sig", newline="") as f:
    writer = csv.writer(f)
    
    # 헤더 쓰기
    writer.writerow(columns)
    
    # 데이터 작성
    for item in data['body']['items'][:10]:
        writer.writerow([
            item['ctprvnCd'],
            item['ctprvnNm'],
            item['signguCd'],
            item['signguNm'],
            item['adongCd'],
            item['adongNm'],
            item['stdrDt']
        ])
```
